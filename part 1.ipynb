```python
teams = df.HomeTeam.unique()
```

```python
# full time home team goals versus expected goals
fig = plt.figure(figsize=(8,6))
ax = fig.add_axes([0, 0, 1, 1])
colors = plt.get_cmap('PuOr', 2)
actual_pts={}
expected_pts={}

for team in teams:
    actual_pts[team]=df[df["HomeTeam"]==team].FTHG.sum()
    expected_pts[team]=df[df["HomeTeam"]==team].HS.sum()

pts_difference = {team: actual_pts[team] - expected_pts[team] for team in teams}
sorted_teams = sorted(pts_difference.keys(), key=pts_difference.get)


# Create data points and connect them with a dashed line
yval = 0
for team in sorted_teams:
    if actual_pts[team] > expected_pts[team]:
        fill_color = colors(0)
    else:
        fill_color = colors(1)

    ax.plot([actual_pts[team], expected_pts[team]], [yval, yval], linestyle='--', linewidth=2, color=fill_color)
    ax.plot(actual_pts[team], yval, 'o', color=fill_color, markersize=10)
    ax.plot(expected_pts[team], yval, 'o', color='white', markeredgecolor=fill_color, markeredgewidth=2, markersize=10)
    yval += 1
    
ax.grid(color='white', linewidth=2)
# Points labels
ax.set_xticks(np.arange(0, 1100, 100))
# Team labels
ax.set_yticks(np.arange(0, len(sorted_teams), 1))
ax.set_yticklabels(sorted_teams)
# Set axis limits
ax.set_xlim(0, 1000)
# Set axis labels
ax.set_xlabel('Points')   
```

```python
teams = df.HomeTeam.unique()
```

```python
#total average of all teams

totalaverage={}

for team in teams:
    df1=df[df["HomeTeam"]==team]
    df2=df[df["AwayTeam"]==team]
    a=pd.concat([df1,df2]).shape[0]
    totalaverage[team]=((df1.FTHG.sum())+(df2.FTAG.sum()))/a
```

```python
totalaverage  #average of all teams as hometeam and awayteam
```

```python
#average of number of hometeam and awayteam 
hometeamshots={}
awayteamshots={}

for team in teams:
    hometeamshots[team]=((df[df["HomeTeam"]==team].FTHG.mean()))
    awayteamshots[team]=((df[df["AwayTeam"]==team].FTAG.mean()))
```
